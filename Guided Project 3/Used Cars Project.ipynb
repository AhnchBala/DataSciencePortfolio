```python
regn_year_bool = (autos['registration_year'] <1900) | (autos['registration_year'] >2016)
autos.drop(autos[regn_year_bool].index, inplace=True)
```

Earliest Car was from 1910.

```python
brand_dict = {}
```

```python
brands = autos['brand'].value_counts(normalize=True)
```

```python
brands,len(brands)
```

```python
top_brands = brands.head(20).index
top_brands 
```

We have chosen the top 20 brands of cars being sold based on total count. Since this is Germany, VW is the highest, with bmw coming in second.

```python
for b in top_brands:
    # select all the rows of that brand
    selected_rows = autos[autos['brand']==b]
    # get the average price
    avg_price = int(selected_rows['price'].mean())
     # Assign the mean value in dictionary
    brand_dict[b] = avg_price
```

```python
brand_dict
```

Sonstige Autos has the highest price in the list of top 20. VW has a decent price of 5402.

```python
avg_brand_mileage ={}
for b in top_brands:
    # select all the rows of that brand
    selected_rows = autos[autos['brand']==b]
    # get the average price
    avg_mileage = int(selected_rows['odometer_km'].mean())
     # Assign the mean value in dictionary
    avg_brand_mileage[b] = avg_mileage
```

```python
avg_brand_mileage
```

```python
bmp_series=pd.Series(brand_dict).sort_values(ascending=False)
bmm_series=pd.Series(avg_brand_mileage)
```

```python
top_20_cars_sold = pd.DataFrame(bmp_series,columns=['mean_price'])
```

```python
top_20_cars_sold['avg_brand_mileage_km'] = pd.DataFrame(bmm_series)
```
